### Anagrams Game: Build a Scrambled Word Game using Pythonüêç

```python
import random 

words = ["iranman","harry","thor","wanda","jin","salman","shahrukh","awais"]

word = random.choice(words)
jumble = "".join(random.sample(word,len(word)))

chances = 5
print("~"*40)
print("~~~~~~~~~~~ ANAGRAMS GAMEüëªüëª ~~~~~~~~~~~~")
print("~"*40)
print("~"*40)

while chances != 0:
    print(f"THE WORD IS {jumble}")
    
    guess = input("enter your guess jumble here ").lower()
    if guess ==  word:
        print("YOUR GUESS IS CORRECT ")
        
        print("YOU WON üòçüëå ")
        break
    else:
        chances -=1
        print("INCORRECT GUESS REMAINING GUESS ARE ",chances)
else:
    print("ALL YOUR CHANCES IS FINISH NOW ")
    
    print("YOU  LOSS THE GAME ü§∑‚Äç‚ôÄÔ∏èü§î")
    
    print("THE CORRECT WORD IS ",word)
    
print("THANKS FOR PLAYING ")
    
    
```
