## To-Do List Manager (File I/O)

```python
to_do = []

def load_tasks():
    global to_do
    try:
        with open("todo.txt" , "r") as file:
            for line in file :
                to_do.append(line.strip)
        print("DO_TO list loaded succesfully ")
        
    except FileNotFoundError:
        print("to_do list not found starting with the empty list ")
        
load_tasks()

def save_tasks():
    try:
        with open("to_do.txt " , "w") as file :
            for task in to_do :
                file.write(task + '\n')
        print("to_do list save successfully ")
        
    except Exception as e:
        print(f"the the error while saving the file {e}")
        
def display_menu():
    
    """~~~~~~~~~~Prints the main menu options to the user.~~~~~~~~"""
    print("\n--- To-Do List Manager ---")
    print("1. View To-Do List")
    print("2. Add New Task")
    print("3. Delete Task")
    print("4. Save and Quit")
    print("--------------------------")
    
def main_loop():
    """~~~~~~Contains the main execution loop for the To-Do List program.~~~~~~"""
    
    load_tasks()
    
    while True:
        #  Call display_menu()
        display_menu()
        
        choice = input("Enter your choice (1-4): ").strip()
        
        if choice == '4':
            save_tasks() 
            print("To-Do List saved. Goodbye!")
            break
        
        elif choice in ('1', '2', '3'):
            print(f"Option {choice} selected. Logic coming soon!")
        else:
            print("Invalid choice. Please enter a number between 1 and 4.")

def view_tasks():
    """Displays the current to-do list with numbered indices."""
    global todo_list
    if not todo_list:
        print("\n Your to-do list is empty! Add a task.")
        return

    print("\n--- YOUR TO-DO LIST ---")
    for i, task in enumerate(todo_list):
        # We start numbering at 1 for user readability (i + 1)
        print(f"{i + 1}. {task}")
    print("-----------------------")


def add_task():
    """Prompts the user for a new task and adds it to the list."""
    global todo_list
    new_task = input("Enter the new task to add: ").strip()
    
    if new_task:
        todo_list.append(new_task)
        print(f" Task '{new_task}' added.")
    else:
        print(" Task cannot be empty.")


def delete_task():
    """Prompts the user for a task number and removes it from the list."""
    global todo_list
    
    if not todo_list:
        print("\nNothing to delete! The list is empty.")
        return
        
    view_tasks()
    
    try:
        task_num_str = input("Enter the number of the task to delete: ").strip()
        task_index = int(task_num_str)
        
        if 0 <= task_index < len(todo_list):
            deleted_task = todo_list.pop(task_index)
            print(f"ðŸ—‘ï¸ Task '{deleted_task}' (No. {task_index + 1}) deleted successfully.")
        else:
            print(" Invalid task number. Please choose a number from the list.")
            
    except ValueError:
        print("Invalid input. Please enter a whole number.")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")
        
def main_loop():
    """Contains the main execution loop for the To-Do List program."""
    
    load_tasks() 
    
    while True:
        display_menu()
        
        choice = input("Enter your choice (1-4): ").strip()
        
        if choice == '1':
            view_tasks()
            
        elif choice == '2':
            add_task()
            
        elif choice == '3':
            delete_task()
            
        elif choice == '4':
            save_tasks()
            print("To-Do List saved. Goodbye!")
            break
            
        else:
            print(" Invalid choice. Please enter a number between 1 and 4.")
```
